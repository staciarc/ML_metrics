 6. Функция
 
Напишите функцию, которая принимает на вход список разных моделей и соответствующий модели список параметров в виде словаря. Функция возвращает словарь, который содержит все 4 метрики, расмотренных в этом задании, для каждой модели из списка. 
{'model1': {'accuracy': 0.7, 'precision': 0.5, ... },'model2': {...}}

```python
def metrics(estimaters, params, X_train, y_train, X_test, y_test):
    result = dict()
    for i in range(len(estimaters)):
        estimater = estimaters[i].set_params(**params[i])
        estimater.fit(X_train, y_train)
        y_pred = estimater.predict(X_test)
        y_pred_proba = estimater.predict_proba(X_test)
        result['model' + str(i + 1)] = dict()
        result['model' + str(i + 1)]['accuracy'] = accuracy_score(y_test, y_pred)
        result['model' + str(i + 1)]['precision'] = precision_score(y_test, y_pred, average = "weighted")
        result['model' + str(i + 1)]['recall'] = recall_score(y_test, y_pred, average = "weighted")
        result['model' + str(i + 1)]['roc_auc'] = roc_auc_score(y_test, y_pred_proba, average = "weighted", multi_class = "ovo")
    return result
```

```python
metrics([SVC(), DecisionTreeClassifier(), RandomForestClassifier()],
        [{'C' : 10,
         'class_weight' : None,
         'gamma' : 'auto',
         'kernel' : 'rbf',
         'random_state' : 21,
         'probability' : True},
        {'class_weight' : 'balanced',
        'criterion' : 'gini',
        'max_depth' : 21,
        'random_state' : 21},
        {'class_weight' : 'balanced',
        'criterion' : 'entropy',
        'max_depth' : 24,
         'n_estimators' : 100,
        'random_state' : 21}],
        X_train, y_train, X_test, y_test)
```
